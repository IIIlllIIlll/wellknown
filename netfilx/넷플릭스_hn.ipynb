```python
n_data = data.drop(['title', 'director','cast','description'], axis = 1)
n_data = n_data.dropna()
n_data
```

```python
plt.figure(figsize=(15, 7))
labels=['TV Show', 'Movie']
plt.pie(n_data['type'].value_counts().sort_values(),labels=labels,
        autopct='%1.2f%%', startangle=90)
plt.show()
```

```python
n_data["country"] = n_data["country"].fillna("unknown")
n_data["country"].isnull().sum()
```

```python
n_data["country"].unique()
```

```python
n_data["n_country"]  = n_data["country"].apply(lambda x : x.split(",")[0])
n_data.head(1)
```

```python
plt.figure(figsize =(22,5)) 
plt.xticks(rotation = 90)
sns.countplot(data=n_data, x = 'n_country', hue = 'type')
```

```python
n_data["n_year"]  = n_data["date_added"].apply(lambda x : x.split(",")[-1])
n_data["n_month"]  = n_data["date_added"].apply(lambda x : x.split(" ")[0])

n_data.head()
```

```python
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'Malgun Gothic' in f.name]
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize =(15,5)) 
plt.xticks(rotation = 90)
plt.title('연도별 작품 수', color = 'b', size = 20, pad =10)
sns.countplot(data = n_data, x =  n_data["n_year"])
```
